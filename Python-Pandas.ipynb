# Merge & Join in Pandas 

```python
customerDf = pd.read_csv('Dataset/Ecommerce/customers_dataset.csv')
ordersDf = pd.read_csv('Dataset/Ecommerce/orders_dataset.csv')
```

```python
customerDf.merge(ordersDf, left_on= 'customer_id', right_on= 'customer_id', how = 'inner')
```

```python
customerDf.join(ordersDf,lsuffix ='_')
```

# Concat in Pandas
